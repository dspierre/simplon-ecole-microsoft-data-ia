```
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```
from google import drive
drive.mount('content/drive')
merge_df=pd.read_csv('/content/drive/My Drive/GlobalLandTemperaturesByCity.csv')
```

```
merge_df=pd.read_csv('/content/sample_data/GlobalLandTemperaturesByCity.csv', sep=',')
merge_df.head(5)
```

RENNES

```
df_rennes= merge_df.loc[merge_df.City =='Rennes']
df_rennes
```

```
median = df_rennes['AverageTemperature'].median()
df_rennes['AverageTemperature'].fillna(median, inplace=True)
median =df_rennes['AverageTemperatureUncertainty'].median()
df_rennes['AverageTemperatureUncertainty'].fillna(median, inplace=True)
df_rennes
df_rennes.isnull().sum()
```

```
df_rennes.dtypes
```

```
df_rennes['dt'] = pd.DatetimeIndex(df_rennes['dt'])
df_rennes.dtypes
```

```
df_rennes = df_rennes.rename(columns={'dt': 'ds','y': 'y'})

df_rennes.head(5)
```

```
ax0 = df_rennes.set_index('ds').plot(figsize=(12, 8))
ax0.set_ylabel('')
ax0.set_xlabel('Date')
plt.show()
```

```
my_model = Prophet(interval_width=0.85)
my_model
```

```
m0 = Prophet()
m0.fit(df_rennes)
```

```
futur0 = m0.make_futur0_dataframe(periods=360)
futur0.tail()
```

```
forecast0 = m0.predict(futur0)
forecast0[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```
fig1 = m0.plot(forecast0)
```

```
fig2 = m0.plot_components(forecast0)
```

```
from fbprophet.plot import plot_plotly, plot_components_plotly
plot_plotly(m0, forecast0)
```

```
plot_components_plotly(m0, forecast0)
```
